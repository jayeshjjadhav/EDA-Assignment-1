```python
import pandas as pd
import numpy as np
#replace the csv file with the required path
df = pd.read_csv(r"nutrients_csvfile - nutrients_csvfile.csv")
#cleaning data(missing values=)
cleaning_values = ["t","t'","a"]
df = df.replace(cleaning_values, 0)
df =df.replace(",","", regex=True)
df['Calories'][91]=(8+44)/2

#converting string datatype to int/float
df[["Grams","Protein","Fat"]]=df[["Grams","Protein","Fat"]].astype('int64')
df[["Calories","Sat.Fat","Fiber","Carbs"]]=df[["Calories","Sat.Fat","Fiber","Carbs"]].astype('float64')

#top five protien rich food
rich_protein = df.sort_values(['Protein'], ascending=[0])
rich_protein = rich_protein[["Food","Protein"]].head(5)
print("Top five protien rich food\n",rich_protein)
print('* '*50 + "\n")
#top five high caloric food to be avoided
high_caloric = df.sort_values(["Calories"], ascending=[0])
high_caloric = high_caloric[["Food", "Calories"]].head(5)
print("Top five high caloric food to be avoided\n",high_caloric)
print('* '*50 + "\n")

#Average nutrient content in each food group
print("Average nutrient content in each food group : ")
average_nutrients=df.groupby("Category").mean().round(2)
print(average_nutrients)
print('* '*50 + "\n")

#foods very low in saturated fats
low_sat_fat = df[df["Sat.Fat"] > 0.0]
low_sat_fat = low_sat_fat.sort_values(["Sat.Fat"], ascending=[1])
low_sat_fat = low_sat_fat[["Food","Sat.Fat"]].head(5)
print("Foods very low in saturated fats\n",low_sat_fat)
```
